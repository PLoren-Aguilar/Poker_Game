```python
import pygame
import sys
import random as rnd
import string
import include.player as player
import include.table as table
import include.menu as menu
from include.treys.treys import Card
from include.treys.treys import Evaluator
from include.treys.treys import Deck

pygame.init()

#Create the players
nplayers = 4
players  = []
croupier = table.newtable(nplayers,players)
        
#Create game window and title
screen = menu.newmenu()
surface = pygame.display.set_mode((1200, 1200))
surface.fill((255,255,255))

#Initialise cards' tools
card = Card()

endgame = False
while (endgame == False):
    screen.create_table(surface)
    screen.put_all_cards(surface,len(croupier.players_order),players)
    screen.put_buttons(surface,croupier.players_order,1)
    screen.remove_chips(surface)

    #Initialise the deck of cards
    deck = Deck()
    
    #Give every player a hand
    print(players)
    croupier.set_hand(players,deck)

    #Reveal player's hand
    if (players[0].cash > 0):
        screen.player_cards(surface,players[0].cards,0)

    #Put the blinds for this round
    screen.show_blinds(surface)
    screen.put_blinds(surface,croupier.players_order)
    
    #Put the common cards on the table
    croupier.set_common(deck)
    pygame.display.update()
    
    #Pre-flop
    croupier.set_bet('preflop')
    croupier.bet_round(screen,surface,card,players,'preflop')
    
    #Flop
    croupier.set_bet('postflop')
    croupier.bet_round(screen,surface,card,players,'flop')

    #Turn
    croupier.set_bet('postflop')
    croupier.bet_round(screen,surface,card,players,'turn')

    #River
    croupier.set_bet('postflop')
    croupier.bet_round(screen,surface,card,players,'river')
    
    #Check Winner
    croupier.check_winner(screen,surface,players,card)
    
    #players[winner].cash += croupier.pot  
    croupier.pot = 0
    
    #Check if any player has been eliminated and reactivate any active but folded player
    for indx,player in enumerate(players):
        if (player.cash == 0):
            pass
            #player.active = False
        else:
            if (player.folded == True):
                player.folded = False        
      
    #Clean the table and prepare for a new round 
    surface = pygame.display.set_mode((1200, 1200)) 
    surface.fill((255,255,255))
    croupier.set_dealer(players)

pygame.quit()
sys.exit()
```
